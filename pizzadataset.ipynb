```python
# Convert 'order_date' to datetime with format specified
dataframe['order_date'] = pd.to_datetime(dataframe['order_date'], format='%d-%m-%Y')

# Add a new column 'month'
dataframe['month'] = dataframe['order_date'].dt.month
```

```python
dataframe
```

```python
dataframe['month_text']=dataframe['order_date'].dt.month_name().str.slice(stop=3)
```

```python
dataframe
```

```python
dataframe.info()
```

```python
import matplotlib.pyplot as plt

# Assuming 'total_price' is the column containing the total price
plt.bar(dataframe['month_text'], dataframe['total_price'])
plt.xlabel('Month')
plt.ylabel('Total Price')
plt.title('Total Price by Month')

# Find the index of the maximum total price
max_index = dataframe['total_price'].idxmax()

# Set the color of the bar with the maximum total price to red
bars = plt.bar(dataframe['month_text'], dataframe['total_price'])
bars[max_index].set_color('red')

plt.show()
```

```python
import matplotlib.pyplot as plt
# Assuming 'total_price' is the column containing the total price
plt.bar(dataframe['pizza_category'], dataframe['total_price'])
plt.xlabel('pizza category')
plt.ylabel('Total Price')
plt.title('Total sales by pizza category')
plt.show()
```
